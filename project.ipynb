```python
# SESSION 0PLUS TO 10 
 
#imports my google drive into colab
from google.colab import drive
drive.mount('/content/gdrive')

from matplotlib.colors import ListedColormap
from sklearn.neighbors import kneighbors_graph

#import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.cluster import SpectralClustering


#uses pandas read_csv function to read csv and store in variable; sets the engine to python 
session010 = pd.read_csv('gdrive/My Drive/sessions_0plus_to_10_metadata.csv', engine='python')

#extract specific columns and store in variable df (dataframe)
df = pd.DataFrame(session010, columns=['question1', 'question2','likes', 'shared media', 'follows', 'followed_by', 'cyberaggression', 'cyberbullying']
)

#convert data types of q1 to binary (0 or 1)
#agression and bullying instances set to 1
df['question1'] = (df['question1'] == 'aggression').mul(1)
df['question2'] = (df['question2'] == 'bullying').mul(1)

#takes out string from column and changes the type from string to int
df['likes'] = df['likes'].map(lambda x: x.rstrip('\r\r likes')).astype(int)

print("--------------------------------------------------------------------------------------------------")
print("\t\t\t\t\tThis is session_0plus_to_10")

#convert df to numpy array
df.to_numpy()

df
```

```python
#build a list from df columns
cols = list(df.columns)
df[cols]

#for loop that runs z-score on each column and prints it in new column with column name folowed by zscore
for col in cols:
    col_zscore = col + '_zscore'
    df[col_zscore] = (df[col] - df[col].mean())/df[col].std(ddof=0)

df
```

```python
#knn graph
A = kneighbors_graph(df, 5, mode='connectivity', include_self=True).toarray()
A

#clustering = SpectralClustering(n_clusters=5, affinity='precomputed', n_init=100, assign_labels='discretize')
#clustering.fit_predict(A)
```
